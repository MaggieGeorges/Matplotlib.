## Matplotlib example workflow

```python
# 0. Import matplotlib amd get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot - creating a figure, an axis
fig, ax = plt.subplots(figsize=(10, 10)) #(Width, height)
#10 is the width, 10 is the height ie the dimensions

# 3. Plot data - adding data to our axis
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simmple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (Saving the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a  bar plot from a dictionary
#Creating dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter":  12}
#Creating a figure and axis
fig, ax = plt.subplots()
#pass it keys and values
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) 
#Customizing adding title
ax.set(title="Dan's Nut Butter store",
       ylabel="Price ($)");
```

```python
#Histogram
#Make some data and plot it.
x = np.random.randn(1000)
fig, ax=plt.subplots()
ax.hist(x);
```

.randn returns a sample(s) from a standard normal distribution.
A normal distribution a kind of cemetric curve difference between  mean and variance.
Makes a thousand random samples that will fit in the curve.

```python
#To get multiple plots happenning at the same time-subplots
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

Tuple of a tuple of axis ((ax1, ax2), (ax3, ax4))- are the names of subplots.
It creates a figure and a set of subplots. 
nrows & ncols are the number of rows/columns of the subplot grid.
Now this is a subplot with 4 different subplots.
subplots.plots() can create single plots or multiple using this command.

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
#Plot to eaxh different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

List indexing
ax[0, 0] - row 0, column 0 axis row 0 and plot column 0
ax[0, 1] - row 0, column 1
ax[1, 0] - row 1, column 0
ax[1, 1] - row 1, column 1
Results are similar as using option 1

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range("1/1/2020", periods=1000))
ts
```

We plot the above data after the cumulative sum. Remember the graph will keep changing since we are using random.rand.


```python
# We make something similar to the above nut using the car-sales data.
car_sales
```

```python
# Formatting the price column to integer format
# str.replace- gets the dolar sign, comma and dot and replace it
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
#To get the type of Price column
type(car_sales["Price"][0])
```

```python
# Remove last two zeros
# We use slicing to remove the last 2.
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
#Adding a Sale Date column using the function pd.date_range
#periods-how many dates we add, ie the len of our car_sales
car_sales["Sales Date"]=pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
#Making a total column, we us cumsum() function coz it adds the sales cumulatively
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
#The price column is still in string column it just concated the string together which is not what we want
type(car_sales["Price"][0])
```

```python
#To fix it, we change the string price column to an integer and then call the cumsum()
#.astype(int)-reasigns the total sales column to be the price column as an integer as type int then we call the cumsum() function
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plotting the total sales(Line plot)
# Calling .plot() directly on a pandas df,a pandas version of the matplotlib API/ Pandas rapa for the matplotlb API
car_sales.plot(x="Sales Date", y="Total Sales");
```

```python
#We now plot a scatter plot
#First reassign price column to int
car_sales["Price"]=car_sales["Price"].astype(int)

#Plot scatter plot with price column as numeric
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

### Pyplot vs matplotlib OO method

* When plotting something quickly, its okay to use pyplot method but if advanced, use matplotlib oo method

```python
#Visualizing our data
heart_disease.head()
```

```python
#Creating a new dataframe
#oOf patients over age 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#To get the length of the over_50 df
len(over_50)
```

```python
#Pyplot method-AKA plotting directly from a pandas df with plot.
#We plot a scatter plot of age and cholestral and color it(c) using the target column.
over_50.plot(kind='scatter',
            x='age',
            y='chol', 
            c='target');
#c gives the color column
```

```python
# OO method mixed with pyplot method
fix, ax=plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#To adjust the axis we use:(uncomment to see ow it works)
#ax.set_xlim([45, 100])
#Changes the limitations of our x-axis to be 45-100 instead of 50-80
```

When using the OO method with pandas, the plot function gets the ax parameter that says we are using the OO method and wants it to plot data on the ax parameter created.
Better coz its bigger we have put the fig size parameter, and parsed the ax method and gotten the age column which has now been labelled. 

### Object Oriented(OO) method from scratch

```python
fig, ax = plt.subplots(figsize=(10, 6))

#Plot data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

#Customize the plot- title, labelling the x and y axis.
ax.set(title="Heart Disease and Cholestral levels",
     xlabel="Age",
     ylabel="Cholestral");

#add the legend(for us to know what the dots are)
#legend method, the star npacks the legends elements(from the c parameter) and create a legend off from those elements.
ax.legend(*scatter.legend_elements(), title="Target");

#To add a horizontal line across the axis 
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

The horizontal line indicates the average cholestral.
linestyle indicates it should be dotted.
Signature: ax.hline(y=0, xmin=0, **kwargs)
    we give it y as the cholestral column, the mean of that column.
    **kwargs means keywords arguments dont necessarily appear.
    

```python
over_50.head()
```

```python
#A subplot of cholestral $ thelach on the y-axis and age as their x-axis
#Using oo method from scratch
#Creating a single figure with two axis
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10, 10),
                             sharex=True)

#Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

#Customize ax0
ax0.set(title="Heart Disease and  Cholestral Levels",
        ylabel="Cholestral")

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
#Adding a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");


#Adding data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#Customize a1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");
#Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
#Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Adding a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

### Customizing the y and x axis limitations

```python
#Creating a single figure with two axis
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10, 10),
                             sharex=True)

#Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease and  Cholestral Levels",
        ylabel="Cholestral")

#Change the x axis limit
ax0.set_xlim([50, 80])

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
#Adding a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");


#Adding data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
#Customize a1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");

#Change ax1 x axis limits 
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
#Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
#Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Adding a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
